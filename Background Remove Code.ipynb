```python
import os
from rembg import new_session, remove
from tqdm import tqdm
from PIL import Image
import io

INPUT_ROOT = r"D:\Downloaded FIle\Compressed\AJLDC\AJLDC"   
OUTPUT_ROOT = r"D:\Downloaded FIle\Compressed\AJLDC\AJLDC_BGRemove"               

# Create a session for background removal
session = new_session("u2net")  

SUPPORTED_FORMATS = ('.jpg', '.jpeg', '.png')


os.makedirs(OUTPUT_ROOT, exist_ok=True)

for class_folder in os.listdir(INPUT_ROOT):
    input_class_path = os.path.join(INPUT_ROOT, class_folder)
    output_class_path = os.path.join(OUTPUT_ROOT, class_folder)

    if not os.path.isdir(input_class_path):
        continue

    os.makedirs(output_class_path, exist_ok=True)
    print(f"üîÑ Processing class: {class_folder}")

    # Process all images
    for filename in tqdm(os.listdir(input_class_path)):
        if not filename.lower().endswith(SUPPORTED_FORMATS):
            continue

        input_path = os.path.join(input_class_path, filename)
        output_path = os.path.join(output_class_path, os.path.splitext(filename)[0] + ".png")
        try:
            with open(input_path, "rb") as f:
                input_image = f.read()
            output_image_bytes = remove(input_image, session=session)
            image = Image.open(io.BytesIO(output_image_bytes)).convert("RGBA")
            bg = Image.new("RGBA", image.size, (255, 255, 255, 255))
            image = Image.alpha_composite(bg, image)

            image.save(output_path)

        except Exception as e:
            print(f"‚ùå Error processing {filename} in {class_folder}: {e}")

print("‚úÖ All advanced background removal completed!")
```
