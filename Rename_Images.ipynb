```python
import os
import shutil

source_main = r"D:\Downloaded FIle\Compressed\AJLDC\AJLDC"  
target_main = r"D:\Downloaded FIle\Compressed\AJLDC\AJLDCReneame"  
os.makedirs(target_main, exist_ok=True)

def rename_and_copy_images_from_subfolders(source_main, target_main):

    for subfolder_path, subfolders, files in os.walk(source_main):
        images = [file for file in files
                  if file.lower().endswith(('.jpg', '.jpeg', '.png', '.gif', '.bmp'))]
        if images:
            subfolder = os.path.basename(subfolder_path)
            print(f"Processing folder: {subfolder}")
            target_subfolder = os.path.join(target_main, subfolder)
            os.makedirs(target_subfolder, exist_ok=True)
            for i, filename in enumerate(images):
                source_path = os.path.join(subfolder_path, filename)
                new_filename = f"{subfolder}_{i + 1}.jpg"
                target_path = os.path.join(target_subfolder, new_filename)

                shutil.copy(source_path, target_path)
                print(f"Copied and renamed {filename} to {new_filename} in folder {subfolder}")
rename_and_copy_images_from_subfolders(source_main, target_main)
```
